```python
# import library
import pandas as pd
```

```python
# variables
rdtnd = [94, 77, 49, 45, 43, 32, 49] # random digit for types of news day
tnd = [] # types of news day
rdd = [80, 20, 25, 88, 98, 65, 86] # random digit for demand
dem = [] # demand
rfs = [] # Revenue from sales
lp = [] # Lost profit
sfs = [] # Salvage from scrap
dp = [] # Daily profit
ls = []
stock = int(input("Enter the stock: "))
sp = float(input("Enter the selling price(in cents): "))
bp = float(input("Enter the buying price(in cents): "))
spp = float(input("Enter the scrap per paper(in cents): "))
```

```python
# Logics
# convert sp & bp value to dollar
sp = sp/100.0
bp = bp/100.0
spp = spp/100.0

# 1. tnd
def ltnd(rval):
    if rval <= 35:
        return "Good"
    elif rval <= 80:
        return "Fair"
    else:
        return "Poor"
# 2. demand
def ldem(tval, dval):
    if tval == "Good":
        if dval <= 3:
            return 40
        elif dval <= 8:
            return 50
        elif dval <= 23:
            return 60
        elif dval <= 43:
            return 70
        elif dval <= 78:
            return 80
        elif dval <= 93:
            return 90
        elif dval <= 100:
            return 100
        else:
            return 100
    elif tval == "Fair":
        if dval <= 10:
            return 40
        elif dval <= 28:
            return 50
        elif dval <= 68:
            return 60
        elif dval <= 88:
            return 70
        elif dval <= 96:
            return 80
        elif dval <= 100:
            return 90
        else:
            return 90
    elif tval == "Poor":
        if dval <= 44:
            return 40
        elif dval <= 66:
            return 50
        elif dval <= 82:
            return 60
        elif dval <= 94:
            return 70
        elif dval <= 100:
            return 80
        else:
            return 80
# 3. rfs
def lrfs(d):
    return round(d*sp, 2)

# 4. lp
def llp(d1):
    if d1 <= stock:
        return 0
    else:
        return round((d1-stock)*(sp-bp), 2)
# 5. sfs
def lsfs(d2):
    if d2 <= stock:
        return round((stock-d2)*spp, 2)
    else:
        return 0

# 6. dp
def ldp(r, l, s):
    return (round((r-(stock*bp)-l)+s, 2))
```

```python
for i in range(7):
    tnd.append(ltnd(rdtnd[i]))
    dem.append(ldem(tnd[i], rdd[i]))
    rfs.append(lrfs(dem[i]))
    lp.append(llp(dem[i]))
    sfs.append(lsfs(dem[i]))
    dp.append(ldp(rfs[i], lp[i], sfs[i]))

print(f"tnd: {tnd}")
print(f"dem: {dem}")
print(f"rfs: {rfs}")
print(f"lp: {lp}")
print(f"sfs: {sfs}")
print(f"dp: {dp}")
```

```python
# create csv file
cd = 1
for j in range(7):
    df = {}
    df['Day'] = cd
    cd += 1
    df['Random Digit for types of News day'] = rdtnd[j]
    df['Types of News day'] = tnd[j]
    df['Random Digit for demand'] = rdd[j]
    df['Demand'] = dem[j]
    df['Revenue from Sales'] = rfs[j]
    df['Lost Profit'] = lp[j]
    df['Salvage from scrap'] = sfs[j]
    df['Daily Profit'] = dp[j]
    ls.append(df)

table = pd.DataFrame(ls)
table.to_csv("newspaper.csv")

```
